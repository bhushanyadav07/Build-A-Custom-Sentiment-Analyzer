```
import pandas as pd
df = pd.read_csv('/content/movie_review_data_1000.csv')
df.head()
```

```
review = list(df['review'].values)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
review_vect = tfidf.fit_transform(review)
print(review_vect.shape)
```

####OK review vector have 18506 numeric vecotr for each 1000 reviws


```
df_sentiment = pd.DataFrame(review_vect.todense(), columns = ['col_'+ str(i+1) for i in range(review_vect.shape[1])])
df_sentiment['sentiment'] = df['sentiment']
```

```
df_sentiment.tail()
```

As we can see the output shows numeric vector represtation there are 5 row and 18506 columns for each review  and each word in the doc.
